```python
import matplotlib.pyplot as plt
import re
```

```python
def read_data(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
    return [float(line.split()[1]) for line in lines]

def plot_data(data, ylabel, title):
    plt.plot(data, marker='o')
    plt.xlabel('Batch Number')
    plt.ylabel(ylabel)
    plt.title(title)
    plt.grid(True)
    plt.show()

# File paths
iops_file = "/home/yue21/mlndp/ML_example/result/iops.txt"
throughput_file = "/home/yue21/mlndp/ML_example/result/throughput.txt"
latency_file = "/home/yue21/mlndp/ML_example/result/latency.txt"

# Read data from files
iops_data = read_data(iops_file)
throughput_data = read_data(throughput_file)
latency_data = read_data(latency_file)

# Plot IOPS data
plot_data(iops_data, 'IOPS (batch/s)', 'IOPS Plot')

# Plot Throughput data
plot_data(throughput_data, 'Throughput (data/s)', 'Throughput Plot')

# Plot Latency data
plot_data(latency_data, 'Latency (seconds)', 'Latency Plot')
```
